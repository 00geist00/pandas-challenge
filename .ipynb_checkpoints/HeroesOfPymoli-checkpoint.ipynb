```python
##Heroes of Pymoli Data Analysis
#1) Of the 576 players, 84% are male, 14% are females, and 2% are Other/Non-disclosed.
#2) 44.79% of players are between 20-24 years of age, and they spend 46.8% of total spend on items.
#2) The age range of 35-39 has the highest avg. purchase price for an item at $3.60.
```

```python
#dependencies
import pandas as pd
import numpy as np

# File to Load
file_to_load = "purchase_data.csv"

# Read Purchasing File and store into Pandas data frame
purchase_data = pd.read_csv(file_to_load)
```

```python
#examine the data
purchase_data.head()
```

```python
#Player Count - Display the total number of players
#[:,columns] - returns full list of rows
player_info = purchase_data.loc[:,["SN","Age","Gender"]]
#drop_duplicates() because this is a list of items and not people (so will have multiple entries for 1 player)
player_info = player_info.drop_duplicates()
unique_players = player_info["SN"].count()

# Display the total number of players
pd.DataFrame({"Total Players": [unique_players]})
```

```python
#Purchasing Analysis
##Number of Unique Items
##Average Purchase Price
##Total Number of Purchases
##Total Revenue

##Number of Unique Items
item_count = len(purchase_data["Item ID"].unique())
##Average Purchase Price
avg_item_price = purchase_data["Price"].mean()
##Total Number of Purchases
purchase_count = purchase_data["Price"].count()
##Total Revenue
total_rev = purchase_data["Price"].sum()


#create the summary table to store the information
summary_table = pd.DataFrame({"Number of Unique Items": [item_count],
                             "Average Purchase Price": [avg_item_price],
                             "Total Number of Purchases": [purchase_count],
                             "Total Revenue": [total_rev]})

#formatting
summary_table["Average Purchase Price"] = summary_table["Average Purchase Price"].astype(float).map("${:,.2f}".format)
summary_table["Total Revenue"] = summary_table["Total Revenue"].astype(float).map("${:,.2f}".format)

#display table
summary_table
```

```python
#Gender Demographics
##Percentage and Count of Male Players
##Percentage and Count of Female Players
##Percentage and Count of Other / Non-Disclosed

gender_totals = player_info["Gender"].value_counts()
gender_percents = gender_totals/gender_totals.sum()*100
gender_demographics = pd.DataFrame({"# of Players": gender_totals, "% of Players": gender_percents})

#Formatting clean up
gender_percents = gender_percents.astype(float).map("{:,.2f}%".format)
gender_demographics = pd.DataFrame({"# of Players": gender_totals, "% of Players": gender_percents})

#display table
gender_demographics
```

```python
#Purchasing Analysis (Gender)
##Run basic calculations to obtain purchase count, avg. purchase price, avg. purchase total per person etc. by gender
##Create a summary data frame to hold the results
##Optional: give the displayed data cleaner formatting
##Display the summary data frame

#purchase count, 
#avg. purchase price
#avg. purchase total per person
gender_purchase_spend = purchase_data.groupby(['Gender']).sum()['Price'].map("${:,.2f}".format)
gender_purchase_avg = purchase_data.groupby(['Gender']).mean()['Price'].map("${:,.2f}".format)
gender_purchase_unique = purchase_data.groupby(['Gender']).count()['Price']

gender_purchasing_analysis = pd.DataFrame({"Purchase Count": gender_purchase_unique,
                              "Average Purchase Price": gender_purchase_avg,
                              "Total Purchase Value": gender_purchase_spend})

gender_purchasing_analysis
```

```python
#Age DemographicsÂ¶
##Establish bins for ages
##Categorize the existing players using the age bins. Hint: use pd.cut()
##Calculate the numbers and percentages by age group
##Create a summary data frame to hold the results
##Optional: round the percentage column to two decimal points
##Display Age Demographics Table
```

```python
#Establish bins for ages
age_bins = [0, 9.9, 14.99, 19.90, 24.90, 29.90, 34.90, 39.90, 99999]
group_names = ["<10", "10-14", "15-19", "20-24", "25-29", "30-34", "35-39", "40+"]
```

```python
##Categorize the existing players using the age bins. Hint: use pd.cut()
player_info["Age Demographics"]= pd.cut(purchase_data["Age"], age_bins, labels=group_names)
```

```python
##Calculate the numbers and percentages by age group
age_demographics_totals = player_info["Age Demographics"].value_counts()
age_demographics_percents = age_demographics_totals/age_demographics_totals.sum()*100
age_demographics_percents = age_demographics_percents.map("{:,.2f}%".format)

##Create a summary data frame to hold the results
age_demographics = pd.DataFrame({"Total Count": age_demographics_totals, "Percentage of Players": age_demographics_percents})
```

```python
##Display Age Demographics Table
age_demographics.sort_index()
```

```python
#Purchasing Analysis (Age)
##Bin the purchase_data data frame by age
##Run basic calculations to obtain purchase count, avg. purchase price, avg. purchase total per person etc. in the table below
##Create a summary data frame to hold the results
##Optional: give the displayed data cleaner formatting
##Display the summary data frame
```

```python
##Bin the purchase_data data frame by age

purchase_data["Age Ranges"] = pd.cut(purchase_data["Age"], age_bins, labels=group_names)
```

```python
##Run basic calculations to obtain purchase count, avg. purchase price, avg. purchase total per person etc. in the table below

age_purchase_spend = purchase_data.groupby(['Age Ranges']).sum()['Price'].map("${:,.2f}".format)
age_purchase_avg = purchase_data.groupby(['Age Ranges']).mean()['Price'].map("${:,.2f}".format)
age_purchase_unique = purchase_data.groupby(['Age Ranges']).count()['Price']
```

```python
age_purchasing_analysis = pd.DataFrame({"Purchase Count": age_purchase_unique,
                              "Average Purchase Price": age_purchase_avg,
                              "Total Purchase Value": age_purchase_spend})
```

```python
age_purchasing_analysis
```

```python
#Top Spenders
##Run basic calculations to obtain the results in the table below
##Create a summary data frame to hold the results
##Sort the total purchase value column in descending order
##Optional: give the displayed data cleaner formatting
##Display a preview of the summary data frame


user_total = purchase_data.groupby(["SN"]).sum()["Price"].map("${:,.2f}".format)
user_average = purchase_data.groupby(["SN"]).mean()["Price"].map("${:,.2f}".format)
user_count = purchase_data.groupby(["SN"]).count()["Price"]

user_data = pd.DataFrame({"Purchase Count": user_count,
                          "Average Purchase Price": user_average,
                          "Total Purchase Value": user_total
                          })

user_data.sort_values(by=['Purchase Count'],ascending=False).head()
```

```python
#Most Popular Items
##Retrieve the Item ID, Item Name, and Item Price columns
##Group by Item ID and Item Name. Perform calculations to obtain purchase count, item price, and total purchase value
##Create a summary data frame to hold the results
##Sort the purchase count column in descending order
##Optional: give the displayed data cleaner formatting
##Display a preview of the summary data frame
```

```python
##Retrieve the Item ID, Item Name, and Item Price columns

item_data = purchase_data.loc[:,["Item ID", "Item Name", "Price"]]
```

```python
##Group by Item ID and Item Name. Perform calculations to obtain purchase count, item price, and total purchase value

total_item_purchase = item_data.groupby(["Item ID", "Item Name"]).sum()["Price"]
average_item_purchase = item_data.groupby(["Item ID", "Item Name"]).mean()["Price"]
item_count = item_data.groupby(["Item ID", "Item Name"]).count()["Price"]
```

```python
##Create a summary data frame to hold the results
##Sort the purchase count column in descending order

item_data_pd = pd.DataFrame({"Purchase Count": item_count,"Item Price": average_item_purchase, "Total Purchase Value": total_item_purchase})
item_data_pd.sort_values("Purchase Count", ascending=False).head()
```

```python
#Most Profitable Items
##Sort the above table by total purchase value in descending order
##Optional: give the displayed data cleaner formatting
##Display a preview of the data frame

item_data_pd.sort_values("Total Purchase Value", ascending=False).head()
```
